```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("titanic.csv",usecols=['Cabin','Survived'])
df.head()
```

#Probability Ratio Encoding



```python
#1.Probability of Survived based on cabin--categorical features
#2.Prob(died) or not survived=1-pr(survived)
#3.pr(survived)/pr(not survived)
#4.Dictionary to map cabin with probability
#5.Replace with categorical feature
```

```python
df['Cabin'].fillna("Missing",inplace=True)
```

```python
df
```

```python
df['Cabin']=df['Cabin'].astype(str).str[0]
```

```python
df.head()
```

```python
prob_df=df.groupby('Cabin')['Survived'].mean()
```

```python
prob_df=pd.DataFrame(prob_df)
prob_df
```

```python
prob_df['died']=1-prob_df['Survived']
```

```python
prob_df
```

```python
prob_df['prob_ratio']=prob_df['Survived']/prob_df['died']
```

```python
prob_df
```

```python
cabin_encoded=prob_df['prob_ratio'].to_dict()
cabin_encoded
```

```python
df['Cabin_prob_encoded']=df['Cabin'].map(cabin_encoded)
```
